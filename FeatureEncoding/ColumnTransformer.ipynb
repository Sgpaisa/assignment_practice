```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
#tranformer=ColumnTransformer(transformers=[(),()],remainder='passthrough')
transformer=ColumnTransformer(transformers=[('tr1',OrdinalEncoder(categories=[['unknown','primary','secondary','tertiary']]),['Education',] ),('tr2',OneHotEncoder(),['Job','Marital','Default','Housing','Loan','contact','Month','poutcome'])],remainder='passthrough')
```

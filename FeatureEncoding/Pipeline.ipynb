```python
#4 march 2026
import pandas as pd
```

```python
df=pd.read_csv('Bank_Marketing_Original_Data.csv')
```

```python
df.head()
```

```python
#first step -> separate data in multiple data categories
numeric_data=['Age','Balance','Day','Duration','Campaign','Pdays','Previous']

ordinal_data=['Education']

nominal_data=['Job','Marital','Default','Housing','Loan','contact','Month','poutcome']
```

```python
numeric_data
```

```python
# separate pipeline transformer for our separate data
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# since column data is numeric hence no encdoing is not needed only StandardScalar()
numeric_pipeline=Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),('scaling',StandardScaler())])
```

```python
numeric_pipeline
```

```python
ordinal_pipeline=Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),('encoding',OrdinalEncoder(categories=[['unknown','primary','secondary','tertiary']])),('scaling',StandardScaler())])
```

```python
ordinal_pipeline
```

```python
#scanling is not needed as OneHotEncoder convert the data into 0 and 1 hence scaling is not needed.

nominal_pipeline=Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),('encoding',OneHotEncoder(handle_unknown='ignore'))])
```

```python
# third step
"""combine pipeline transformer with thier respective data"""
from sklearn.compose import ColumnTransformer
```

```python
feature=ColumnTransformer(transformers=[('numeric transformation',numeric_pipeline,numeric_data),('ordinal transformation',ordinal_pipeline,ordinal_data),('nominal tranformation',nominal_pipeline,nominal_data)])
```

```python
feature
```

```python
#fourth step- create an model with the respective transformer along with their algorith
from sklearn.linear_model import LogisticRegression
```

```python
model=Pipeline(steps=[('feature',feature),('classifier',LogisticRegression())])
```

```python
model
```

```python
#fifth step-> train test split my data and encoding of my output column data
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[:,:-1]
Y=df['TARGET']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
x_test
```

```python
# final step-> model training and getting prediction from the model
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

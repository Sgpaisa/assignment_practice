```python
# mean median imputation --> fill the missing data with the mean or median of all  rows of that column  
#while going mean values is always advisable for the outliers .
# median --> most fequent value from the given column, if you allrnge all the values from the low to high side--> in between value 
#mode --> most frequent value is mode.

```

```python
df['Heart Rate (BPM)'].describe()
```

```python
df.info()
```

```python
#forward filling or backward filling-> time series data 
# forward filling -> fill the missing value with the previous value
# backward filling -> fill the missing value with the next value    


df[df['Step Count'].isnull()==True]
```

```python
# method --> ffill --> forward filling
#backward filling --> bfill
df['Step Count'].fillna(method='ffill', inplace=True)
```

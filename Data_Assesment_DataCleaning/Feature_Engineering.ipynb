```python
titanic=pd.read_csv('Titanic.csv')
```

```python
df=pd.concat([train,titanic])
```

```python
df['Family Size']=df['SibSp']+ df['Parch']+1
```

```python
df.sample(12)
```

```python
sns.barplot(data=df,x='Family Size',y='Survived')
```

```python
def fam_type(num):
    if num>=1 and num<3:
        return 'Small Family'
    elif num>3 and num<=5:
        return 'Medium Family'
    else:
        return 'Large Family'
```

```python
df['Family Type']=df['Family Size'].apply(fam_type)
```

```python
df.sample(10)
```

```python
df['Family Type'].value_counts()
```

```python
sns.heatmap(pd.crosstab(df['Family Type'],df['Survived'],normalize=True)*100,annot=True,cmap='viridis',linewidth=0.5)
```

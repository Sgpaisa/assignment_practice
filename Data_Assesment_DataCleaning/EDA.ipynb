```python
# steps for EDA
''' categorise the given column as a numerical , categorical , adn mixed data type column
univariate analysis- numerical column or categorical column
numerical column analysis- distribution of the data - histogram,kdeplot
--> outliers -> boxplot()
---> missing value->

categorical column analysis-->distribution of the data->countplot
--> missing values->
Bivariate analysis->numerical numerical -> scatterplot
numerical categorical-> barplot
categorical categorical-> heatmap

EDA -> feature engineering
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic.csv')
```

```python
df.head()
```

```python
#univariate analysis on numerical columns-->
#data distribution of the numerical column-> describe() and plotting -> kdeplot or histogram
#outlier detection -> boxplot()
```

```python
df['Age'].describe()
```

```python
df[df['Age']==0.170000]
```

```python
sns.histplot(data=df,x='Age')
```

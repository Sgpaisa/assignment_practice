```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('D:/visual studio/Module and packages/Matplotlib/student_depression_dataset.csv')#data gathering
```

```python
chocsales=pd.read_csv('D:/visual studio/Module and packages/Matplotlib/Chocolate Sales.csv')
```

```python
df.columns
```

```python
#axis level function-> lineplot()
#style --> style parameter can also add addtional insigh in the graph by changing the plot style
sns.lineplot(data=df,x='Age',y='CGPA',hue='Gender',style='Dietary Habits')
plt.show()
```

```python
#figure level funtion-> sns.relplot(,kind='line'), kind='line',kind='scatterplot'
#hue-> it can group or separate teh information in the plot w.r.t a new column
g=sns.relplot(data=df,x='Age',y='CGPA',kind='line',hue='Gender',style='Dietary Habits')
g.savefig('sachin.png')
plt.show()
```

```python
'''matrix plot - relationship between 2-dimension like a table
 
heatmap -> in every classification machine learning problem statement we use the heatmap.heatmap used color shaded to show the size of number in matrix 2-d table
pivot()-> index,column,value-> gap.pivot(index='',columns='',values='')
in matrix plot , we are not having figure level but axis level functn
heatmap()-> data-> annot->cmap--> coolwarm,viridis
 '''
gap.head()
```

```python
dt=gap[(gap['continent']=='Europe')].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(data=dt,annot=True,cmap='viridis',linewidth=0.5)
plt.show()
```

```python
dt=gap.pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(data=dt)
```
